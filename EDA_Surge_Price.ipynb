Null value count for columns Var1 is too high which makes the column not so impactful so recommended to drop that column.

Customer_Since_Months columns can be filled with the mean values of the column.

In Type_Of_Cab & Confidence_Life_Style_Index columns, consider all the null values as a new cab type "F" & confidence as "D".

```python
# Dropping Var1 column
df.drop("Var1", axis=1, inplace=True)

# filling "Customer_Since_Months" column with the mean of the same column
df['Customer_Since_Months'].fillna(int(df['Customer_Since_Months'].mean()),inplace = True)


# Considering null values as new cab type "F" & confidence as "D"
df["Type_of_Cab"] = df["Type_of_Cab"].fillna("F")
df["Confidence_Life_Style_Index"] = df["Confidence_Life_Style_Index"].fillna("D")
```

When tried to fill 'Life_Style_Index' column with mean values, the count of entire null values being filled is relatively very frequent when compared to other values frequency.

So, tried to find relation between Confidence life style and that of life style index.

```python
sns.violinplot(data=df, x="Confidence_Life_Style_Index", y="Life_Style_Index")
```

```python
sns.relplot(df, x="Confidence_Life_Style_Index", y="Life_Style_Index")
```

It can be observed that the Confidence life style index can also be categorised as the value given for life style index. 

So, Life style index column can be dropped as we have the confidence life style index column replacing that column.

```python
#dropping Life_Style_Index column
df.drop("Life_Style_Index", axis=1, inplace=True)
```
